```python
import sys,os,re
import pandas as pd
Delimiter = '\t'
```

### reformat CHG file to generate a heatmap

```python
InFileName1 = "/blue/soltis/shan158538/Methylation/OutPut/DMR_heatmap/CHG_new/Filtered_sorted_all.txt"
```

```python
InFile1 = pd.read_csv(InFileName1, header=None, sep=Delimiter, names = ['chr', 'pos', 'diff_diploids', 'diff_subgenomes'])
```

```python
InFile1
```

```python
InFile1['chr_pos'] = InFile1['chr'] + '_' + InFile1['pos'].astype(str)
```

```python
InFile1
```

```python
columns_to_keep = ['chr_pos', 'diff_diploids']
InFile1_subset_1 = InFile1[columns_to_keep]
```

```python
# add a new column and specify the cotent
content = 'Tdu_vs_Tpr'
InFile1_subset_1['comparison'] = content
```

```python
InFile1_subset_1 = InFile1_subset_1[['chr_pos', 'comparison', 'diff_diploids']]
```

```python
InFile1_subset_1 = InFile1_subset_1.rename(columns={'diff_diploids': 'diff'})
```

```python
InFile1_subset_1
```

```python
InFile1_subset_2 = InFile1[['chr_pos', 'diff_subgenomes']]
InFile1_subset_2['comparison'] = 'Tms-du_vs_Tms-pr'
InFile1_subset_2 = InFile1_subset_2[['chr_pos', 'comparison', 'diff_subgenomes']]
InFile1_subset_2 = InFile1_subset_2.rename(columns={'diff_subgenomes': 'diff'})
```

```python
InFile1_subset_2
```

```python
concat = pd.concat([InFile1_subset_1, InFile1_subset_2], axis = 0)
```

```python
concat
```

```python
concat.to_csv("/blue/soltis/shan158538/Methylation/OutPut/DMR_heatmap/CHG_new/Filtered_sorted_all_reformatted.txt", sep=Delimiter, header=True, index=False)
```
